```python
row_letter = input("Please enter the last letter (A to Z) for the row you've reached on the 384 wells plate: ").upper()

# Validate the row_letter to ensure it's a single alphabet letter
while not row_letter.isalpha() or len(row_letter) != 1:
    print("Invalid input. Please enter a single letter.")
    row_letter = input("Please enter the letter for the row you've reached: ").upper()

# Ask the user for the column number
column_number = input("Please enter the column number you've reached on the 384 wells plate: ")

# Validate the column_number to ensure it's an integer
while not column_number.isdigit():
    print("Invalid input. Please enter a valid number.")
    column_number = input("Please enter the column number you've reached: ")

# Convert the column number to an integer
column_number = int(column_number)

maximum_row_letter = (ord(row_letter) - ord('A') + 1)*24
print(f"{maximum_row_letter}")
```

```python
multiply_input = input("Are you doing doublet or triplet? (doublet = 2, triplet = 3) : ")
multiply = int(multiply_input)

sample_name_list = []
sample_mapping = {}

for row_value in range(ord("A"), ord(row_letter)+1):
  letter = chr(row_value)
  item = input(f"Please enter letter {letter}'s corresponding sample name: ")
  sample_mapping[letter] = item

  for i in range(multiply):
    sample_name_list.append(item)

list_of_null = [None] * len (sample_name_list)
```

```python
ref_gene = input("Please enter your reference gene's name: ")
total_target_gene = input("Please enter the total amount of your target gene: ")
target_genes = int (total_target_gene)

gene_list = []
gene_list.append(ref_gene)
gene_list.append(ref_gene)


for name in range(target_genes):
  number = int(name)
  target_gene_name = input(f"Please enter the name of your target gene_{number+1} : ")
  gene_list.append(target_gene_name)
  gene_list.append(target_gene_name)

print(f" ")
print(f"your Target_gene_list is : {gene_list}")
```
